### replace method to fill missing values

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
df
```

```python
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
new_df = df.replace({'temp':-99999,'windspeed':-99999,'event':0},np.NaN)
new_df
```

```python
df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
df
```

```python
new_df = df.replace({
                     -99999:np.NaN,
                     'no event':'Sunny'
                    })
new_df
```

```python
df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
new_df = df.replace({-99999:np.NaN,'No Event': 'Sunny'})
new_df
```

### Regex 

```python
df = pd.read_csv(r'D:\DATASCIENCE\weather1-data.csv')
df
new_df = df.replace({'temp':'[A-Za-z]',
                     'windspeed':'[A-Za-z]'},'',regex=True)
new_df
```
