```python
df = pd.concat([ind_weather,us_weather])
df
```

```python
df = pd.concat([ind_weather,us_weather],ignore_index=True)
df
```

```python
df = pd.concat([ind_weather,us_weather],keys=["ind","us"])
df    
```

### merge (similar to joins in sql)

```python
df1 = pd.DataFrame({'city':['new york','chicago','san fransisco','california'],
                          'temp':[21,14,35,28]})
```

```python
df2 = pd.DataFrame({'city':['new york','chicago','orlando'],
                          'humidity':[68,65,75]})
```

```python
df = pd.merge(df1,df2)
df
```

```python
df = pd.merge(df1,df2,on='city',how='inner')
df
```

```python
df = pd.merge(df1,df2,on = 'city',how='outer')
df
```

```python
df = pd.merge(df1,df2,on = 'city',how='left')
df
```

```python
df = pd.merge(df1,df2,on = 'city',how='right')
df
```

```python
df = pd.merge(df1,df2,on = 'city',how='outer',indicator=True)
df
```

```python
df1 = pd.DataFrame({'city':['new york','chicago','orlando'],
                          'temp':[21,14,35],
                          'humidity':[68,65,75]})
df1
```

```python
df2 = pd.DataFrame({'city':['new york','chicago','baltimore','california'],
                          'temp':[21,14,35,28],
                          'humidity':[68,65,75,69]})
df2
```

```python
df = pd.merge(df1,df2,on='city',how='outer')
df
```

```python
df = pd.merge(df1,df2,on='city',how='outer',suffixes=['_left','_right'])
df
```
